### Random Forest 
* finite data..since there are less values..if more values..then linear regression

```python
df
```

```python
X=df.iloc[:,:-1]
y=df.outcome
```

```python
##train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
##model creation
from sklearn.ensemble import RandomForestClassifier  ##this is our new model.
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
##model evalution
from sklearn.metrics import accuracy_score, classification_report
y_pred = model.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
print(classification_report(y_pred, y_test))
```

```python
#save after training the model
import joblib
joblib.dump(model,"random_forest_model.pkl")
```

```python
import gradio as gr
import joblib
import numpy as np

# Load model trained on 7 features
model = joblib.load("random_forest_model.pkl")

# Define prediction function
def predict_car_safety(buying, maint, doors, persons, lug_boot, safety, dummy_class):
    input_data = np.array([[int(buying), int(maint), int(doors), int(persons), int(lug_boot), int(safety), int(dummy_class)]])
    prediction = model.predict(input_data)[0]
    return f"Prediction Outcome: {prediction}"

# Gradio input components for 7 features (last is a dummy placeholder for class)
inputs = [
    gr.Dropdown([0, 1, 2, 3], label="Buying (0:high, 1:low, 2:med, 3:vhigh)"),
    gr.Dropdown([0, 1, 2, 3], label="Maintenance (0:high, 1:low, 2:med, 3:vhigh)"),
    gr.Dropdown([0, 1, 2, 3], label="Doors (0:2, 1:3, 2:4, 3:5more)"),
    gr.Dropdown([0, 1, 2], label="Persons (0:2, 1:4, 2:more)"),
    gr.Dropdown([0, 1, 2], label="Lug Boot Size (0:big, 1:med, 2:small)"),
    gr.Dropdown([0, 1, 2], label="Safety (0:high, 1:low, 2:med)"),
    gr.Dropdown([0, 1, 2, 3], label="Dummy Class Input (0:unacc, 1:acc, 2:good, 3:vgood)")
]

# Launch Gradio app
gr.Interface(
    fn=predict_car_safety,
    inputs=inputs,
    outputs="text",
    title="Car Evaluation (7-Feature Input)",
    description="⚠️ This interface uses 7 inputs as expected by your trained model. The 7th is a placeholder for the class label and is not recommended in production."
).launch()
```
